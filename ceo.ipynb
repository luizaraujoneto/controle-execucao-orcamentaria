```python
import os

# Defina o diretório onde os arquivos estão localizados
diretorio = '.'

# Lista todos os arquivos no diretório que terminam com '.xlsx'
arquivos_xlsx = [f for f in os.listdir(diretorio) if f.endswith('.xlsx')]

# Exibe a lista de arquivos
print(arquivos_xlsx)
```

```python
import pandas as pd

# Replace the file path with the location of your Excel file
file_path = 'arquivo-execucao-orcamentaria-tre-al.xlsx'

# Read the Excel file
df = pd.read_excel(file_path, header=1)

print(df.columns)
```

```python
df = df[df['Mês Lançamento'] == 'OUT/2024']
print(df)
```

```python
UGResponsavel = df[ ['UG Responsável Código', 'UG Responsável Nome']]

UGResponsavel = UGResponsavel.drop_duplicates()

UGResponsavel = UGResponsavel[UGResponsavel['UG Responsável Nome'].str.startswith('UGR -')]

UGResponsavel = UGResponsavel.sort_values(by='UG Responsável Nome')

print(UGResponsavel)
```

```python
filtro = df[df['UG Responsável Nome'].str.contains('Corporativas', case=False) & (df['Mês Lançamento'] == 'OUT/2024')]

# Display the content
print(filtro)
```

```python
resultado = filtro.groupby('Conta Contábil')['Saldo - R$ (Conta Contábil)'].sum()
print(resultado)
```

```python
resultado2 = filtro[filtro['Conta Contábil'].str.contains('EMPENHADO A LIQUIDAR') ].groupby('PI Nome')['Saldo - R$ (Conta Contábil)'].sum()
print(resultado2)
```
